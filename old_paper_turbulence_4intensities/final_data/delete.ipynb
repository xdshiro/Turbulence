```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Create the figure and axis
fig, ax = plt.subplots()

# Set the limits of the x-axis and y-axis
x = np.linspace(0, 2 * np.pi, 1000)  # X-axis values (0 to 2)
ax.set_xlim(0, 2 * np.pi)
ax.set_ylim(-10, 10)

# Initialize an empty line plot
line, = ax.plot(x, np.sin(x))

# Function to update the plot
def update(A):
    y = A * np.sin(x)
    line.set_ydata(y)  # Update the y-data of the line
    ax.set_title(f"A = {A:.2f}")  # Set the title to show the current value of A
    return line,

# Create the animation
A_values = np.linspace(0, 10, 100)  # Vary A from 0 to 10 in 100 frames
ani = FuncAnimation(fig, update, frames=A_values, blit=True, interval=100)

# Save the animation as an MP4 file
ani.save('sin_wave_animation.gif', writer='pillow', fps=10)

# Display the animation in the notebook
from IPython.display import HTML
HTML(ani.to_jshtml())
```
